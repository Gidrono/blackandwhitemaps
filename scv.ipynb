```python
from lxml import etree

def extract_layer(input_svg_path, output_svg_path, layer_id):
    # Parse the input SVG file
    tree = etree.parse(input_svg_path)
    root = tree.getroot()

    # Define the SVG namespace
    ns = {'svg': 'http://www.w3.org/2000/svg'}

    # Find the specific layer by its ID
    layer = root.xpath(f"//svg:path[@id='{layer_id}']", namespaces=ns)

    if not layer:
        print(f"Layer with ID '{layer_id}' not found.")
        return

    # Create a new SVG root element
    new_svg_root = etree.Element('svg', nsmap=root.nsmap)

    # Append the found layer to the new SVG root
    new_svg_root.append(layer[0])

    # Write the new SVG to the output file
    new_tree = etree.ElementTree(new_svg_root)
    new_tree.write(output_svg_path, pretty_print=True, xml_declaration=True, encoding='UTF-8')

# Example usage
input_svg_path = 'input.svg'
output_svg_path = 'output.svg'
layer_id = 'path11196'

extract_layer(input_svg_path, output_svg_path, layer_id)
```
