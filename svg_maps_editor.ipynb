```python
import os
from lxml import etree
from IPython.display import display, SVG

def extract_layer(input_svg_path, output_svg_path, layer_id):
    # Check if the input file exists
    if not os.path.isfile(input_svg_path):
        print(f"Error: File '{input_svg_path}' not found.")
        return

    try:
        # Parse the input SVG file
        tree = etree.parse(input_svg_path)
        root = tree.getroot()

        # Define the SVG namespace
        ns = {'svg': 'http://www.w3.org/2000/svg'}

        # Find the specific layer by its ID
        layer = root.xpath(f"//svg:path[@id='{layer_id}']", namespaces=ns)

        if not layer:
            print(f"Layer with ID '{layer_id}' not found.")
            return

        # Get the bounding box of the layer
        bbox = layer[0].get('d')
        if bbox:
            bbox = layer[0].get('d')
            # Create a new SVG root element with the same dimensions and viewBox
            new_svg_root = etree.Element('svg', nsmap=root.nsmap)
            new_svg_root.set('viewBox', root.get('viewBox'))
            new_svg_root.set('width', root.get('width'))
            new_svg_root.set('height', root.get('height'))

            # Append the found layer to the new SVG root
            new_svg_root.append(layer[0])

            # Write the new SVG to the output file
            new_tree = etree.ElementTree(new_svg_root)
            new_tree.write(output_svg_path, pretty_print=True, xml_declaration=True, encoding='UTF-8')

            print(f"Layer '{layer_id}' successfully extracted to '{output_svg_path}'.")

            # Display the resulting SVG in the Jupyter Notebook
            with open(output_svg_path, 'r') as svg_file:
                svg_content = svg_file.read()
                display(SVG(svg_content))
        else:
            print(f"Bounding box for layer '{layer_id}' not found.")

    except Exception as e:
        print(f"Error processing file: {e}")

# Example usage
input_svg_path = os.path.abspath('Syrian_Civil_War_map.svg')
output_svg_path = os.path.abspath('output.svg')
layer_id = 'path11196'

extract_layer(input_svg_path, output_svg_path, layer_id)
```
